```python
import matplotlib.pyplot as plt

year_income = df.groupby("year")["Annual_Income"].sum()

plt.figure(figsize=(10, 6))
plt.plot(year_income.index, year_income.values, marker='o', label='Total Annual Income')

plt.xlabel("Year")
plt.ylabel("Total Value")
plt.title("Yearly Totals")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

df['year'] = pd.to_datetime(df['Date_Joined']).dt.year

year_purchase = df.groupby("year")["Purchase_Frequency"].sum()
year_score = df.groupby("year")["Spending_Score"].sum()

plt.figure(figsize=(10, 6))
plt.plot(year_purchase.index, year_purchase.values, marker='s', label='Total Purchase Frequency')
plt.plot(year_score.index, year_score.values, marker='^', label='Total Spending Score')

plt.xlabel("Year")
plt.ylabel("Total Value")
plt.title("Yearly Totals:Purchase Frequency, and Spending Score")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
unique_customers = df.groupby("Region")["Customer_ID"].nunique().reset_index()
unique_customers = unique_customers.rename(columns={"Customer_ID": "Unique_Customers"})

sns.barplot(x="Region", y="Unique_Customers", data=unique_customers,
            order=unique_customers.sort_values("Unique_Customers", ascending=False)["Region"])
```

```python
unique_customers_gender = df.groupby("Gender")["Customer_ID"].nunique().reset_index()
unique_customers_gender = unique_customers_gender.rename(columns={"Customer_ID": "Unique_Customers"})

sns.barplot(x="Gender", y="Unique_Customers", data=unique_customers_gender,
            order=unique_customers_gender.sort_values("Unique_Customers", ascending=False)["Gender"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Date_Joined'] = pd.to_datetime(df['Date_Joined'])
df['year'] = df['Date_Joined'].dt.year


plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Spending_Score', data=df, estimator='mean', palette='viridis')
plt.title("Average Spending Score by Region")
plt.xlabel("Region")
plt.ylabel("Average Spending Score")
plt.tight_layout()
plt.savefig("barplot_spending_score_by_region.png", dpi=300)
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='Annual_Income', y='Spending_Score', hue='Gender', data=df, palette='Set2')
plt.title("Income vs Spending Score by Gender")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig("scatter_income_vs_spending.png", dpi=300)
plt.show()
```
